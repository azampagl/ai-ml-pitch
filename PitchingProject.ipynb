In baseball a batter successfully reaches base between two and three out of every ten attempts. Even in the height of the steroid era, the [league-wide batting average](http://thereadzone.com/2014/05/27/a-steroid-era-hangover-the-strikeout-conundrum/) only rose to .270. Since the 1900's the dominance of the pitcher is one of the few things that hasn't changed in the sport (unlike the type of ball, inclusion of the DH, free-agency, and breaking of the color barrier just to name a few things that have changed). Batters commonly move as far back in the batters box as possible to see a pitch longer, look for any indication of which pitch is coming, and even use pine tar to have better control of the bat - all in the hopes reducing the advantage of the pitcher.

With the inclusion of more comprehensive [sabermetric statistics](http://sabr.org/sabermetrics) and tracking technologies like MLB Gameday's [PitchF/X](http://www.fangraphs.com/library/misc/pitch-fx/), there is a greater opportunity to identify weaknesses of pitchers, batters, and positional players than ever before. The motivation of this project is to see how reasonable it is to aggregate all of a pitcher's pitches throughout his career (going back to 2007, the start of PitchF/X technology), in order to expose any tendencies that can be used to close the advantage gap between the batter and pitcher.

Due to the wide range of pitching styles (predominantly fastball-throwers, predominantly off-speed, groundball-inducing pitcher, fly-ball inducing pitcher, starter, middle-relief, closer, etc), we chose to investigate pitching strategies while focusing on one pitcher at a time. The rationale behind this is that we want to show the type of analysis an opposing coach may be able to carry out on the day before a game in order to have a better understanding of the pitcher they will face. A few questions a coach might like to know the answers to beforehand:

  - Does a pitcher's strategy change over the course of the game?

  - What type of pitch can a batter expect on the first pitch of the at-bat?

  - Can you identify a pitcher changing his usual strategy to accomodate to the weaknesses/strengths of a particular batter?

  - Has the pitcher changed strategy over the course of his career? 

We want to keep our analysis very granular, focusing on the analysis we can carry out for one particular pitcher. If we can prove that our analysis gives some intuition for a particular pitcher whom we choose to have a large enough sample of pitches to aggregate, then it serves as proof-of-concept that our analysis can be carried out to other pitchers as well. 

An extra goal if we can find reasonable answers to the aforementioned questions, would be to see how accurately we can predict pitches for a given pitcher, handedness of batter, strikes, balls, outs, and inning. First, we would have to identify which predictors (balls, strikes, outs, handedness of batter, etc) have the greatest effect on the pitch type. One possible approach would be to see if we could apply machine learning techniques like neural networks or SVMs to predict pitches with reasonable accuracy. 

In a sport where an average batter only succeeds 25% of the time, hopefully we could show methods by which the batter can increase his odds of success by being better prepared with knowledge in a given situation.

### Introduction ###
The first step of our project was to obtain our data and post-process it for future analysis. The ideal scenario would be a single CSV file for each pitcher that contained all the respective pitching data. This CSV file could then be easily accessed as a Panda's DataFrame for future analysis.

In order to obtain our data, we crawled and scraped [brooksbaseball.net](http://www.brooksbaseball.net) - a website dedicated for Sabermetricians interested in MLB pitching data. Unfortunately, the website structured the pitching data on a game basis instead of a pitcher basis - i.e., the HTML pages contained pitching data per game, not per pitcher. We decided that the most efficient route would be to crawl every single game on the website, save the raw data to the local file system, and manipulate the raw data into a multiple pitcher-based CSV files.

### Scraping ###

To scrape [brooksbaseball.net](http://www.brooksbaseball.net), we used a tool called [scrapy](http://scrapy.org/) - an open source python web crawling tool. To use scrapy, an end user must define a root URL and the code used to parse the HTML and crawl through the site until desired pages are reached. For our project, we neeeded to obtain the pitching details of every single game from 2008 to 2014. We did this by chronologically going through every single day and "downloading" all the games on that particular day - [an example of raw HTML data for one game can be found here](http://www.brooksbaseball.net/pfxVB/tabdel_expanded.php?pitchSel=502085&game=gid_2014_07_04_nyamlb_minmlb_1/&s_type=&h_size=700&v_size=500).

Below is the core source code snippet used to scrape the site. However, scrapy projects are standalone python executables. The below source code is non-operational inside a ipython nodebook. The full executable and source code can be found [here](https://github.com/azampagl/ai-ml-pitch/tree/master/wrangle/scrape).

```python
import numpy as np
import scipy as sp
import pandas as pd
import sklearn
import seaborn as sns
from matplotlib import pyplot as plt

from sklearn.ensemble import RandomForestClassifier
from sklearn.cross_validation import cross_val_score

from matplotlib.path import Path
from matplotlib.spines import Spine
from matplotlib.projections.polar import PolarAxes
from matplotlib.projections import register_projection

%matplotlib inline
```

```python
"""
Scrapes brooksbaseball and stores the raw html files to disk.

The style guide follows the strict python PEP 8 guidelines.
@see http://www.python.org/dev/peps/pep-0008/

@author Aaron Zampaglione <azampaglione@g.harvard.edu>
@author Fil Piasevoli <fpiasevoli@g.harvard.edu>
@author Lyla Fadden <lylafadden@g.harvard.edu>

@requires Python >=2.7
@copyright 2014
"""
import os

from datetime import date, timedelta
from dateutil.rrule import rrule, DAILY
from urllib import urlencode
from urlparse import parse_qs, urlparse

from scrapy import Request, Spider

from brooksbaseball.items import GameItem

class RootSpider(Spider):
    # Name of the scraper.
    name = "brooksbaseball"

    # The allowed domains to crawl.
    allowed_domains = ["brooksbaseball.net"]

    # The timeout between requests.
    download_delay = 0.5

    # The base url.
    base_url = "http://www.brooksbaseball.net/pfxVB/"

    # The location to save the scraped html files.
    store_path = "data/"

    # Define a start and end date for crawling.
    #  Baseball seasons starts in March!
    start_date = date(2008, 3, 1)
    end_date = date(2014, 12, 01)

    def __init__(self, *args, **kwargs):
        """
        Initialize the scraper.
        """

        super(RootSpider, self).__init__(*args, **kwargs)

        self.start_urls = []

        # The start urls are every single date from the start to the end dates
        #  defined above.
        for dt in rrule(DAILY, dtstart=self.start_date, until=self.end_date):
            url = self.base_url + "pfx.php/?league=mlb&"
            # The site does not like 0 padded dates.
            url += "&year=" + str(int(dt.year))
            url += "&month=" + str(int(dt.month))
            url += "&day=" + str(int(dt.day))
            self.start_urls.append(url)

    def parse(self, response):
        """
        Parses the first level, which is a drop down box
        of all the games for that day.
        """

        # Find the games selection drop down box.
        xpath = '//select[@name="game"]/option'

        # For each option (game), generate a
        #  new request with game details.
        for sel in response.xpath(xpath):
            url = response.url

            params = parse_qs(
                urlparse(url).query,
                keep_blank_values=True)

            game = sel.xpath('@value').extract()[0]

            url = response.url
            url += "&prevDate=" + \
                str(int(params['month'][0])) + \
                str(int(params['day'][0]))
            url += "&prevGame=" + game
            url += "&game=" + game

            # Generate a new request which is specific to a game
            #  on a particular day.
            req = Request(url, callback=self.parse_game)

            yield req

    def parse_game(self, response):
        """
        Parses the second level, which has a dropdown box of all
        the pitchers for that game.
        """

        # Find the pitcher selection drop down box.
        xpath = '//select[@name="pitchSel"]/option'

        # For each option, generate a new request
        #  which will give the details for a particular
        #  pitcher is a particular game.
        for sel in response.xpath(xpath):
            url = response.url

            url += "&pitchSel=" + sel.xpath('@value').extract()[0]

            req = Request(url, callback=self.parse_pitcher)

            yield req

    def parse_pitcher(self, response):
        """
        Parses the third level, which is the pitcher details page
        for a particular game. It contains a link to a raw
        HTML page that contains all the stats.
        """

        params = parse_qs(
            urlparse(response.url).query,
                keep_blank_values=True)

        data = {}
        data['year'] = params['year'][0]
        data['month'] = params['month'][0]
        data['day'] = params['day'][0]
        data['pitcher_id'] = params['pitchSel'][0]
        data['game_id'] = params['game'][0]

        # Find the raw html table link.
        xpath = '//a[text()="Get Expanded Tabled Data"]'

        # Generate the url to open the raw html table page.
        sel = response.xpath(xpath)[0]

        link = urlparse(sel.xpath('@href').extract()[0])
        params = parse_qs(link.query, keep_blank_values=True)

        for key, value in params.items():
            params[key] = value[0]
        # Override the rows/column size just in case.
        params['h_size'] = 3000
        params['v_size'] = 3000

        url = self.base_url + link.path + "?" + urlencode(params)

        req = Request(url, callback=self.parse_pitcher_stats, meta={'data': data})

        yield req

    def parse_pitcher_stats(self, response):
        """
        Parses the final page of the operation - a raw html page
        with one table containing all the stats for one pitcher during
        a particular game. This html page will be directly downloaded to disk
        for post processing.

        The files will be structured as follows:

        root/
          2008/
            03-01/
              game01.html
              game02.html
              ...
            ...
          ...
        """

        data = response.meta['data']

        # Create a top level year directory that will contain
        #  sub-directories for every day containing at least
        #  one game during the year.
        path = self.store_path
        path += "/" + data['year']

        if not os.path.exists(path):
            os.makedirs(path)

        path += "/" + data['month'].zfill(2) + "-" + data['day'].zfill(2)

        if not os.path.exists(path):
            os.makedirs(path)

        # Make sure the game id and the pitcher id are in the file name.
        filename = path + "/" + \
            data['game_id'][:-1] + \
            "-pid_" + data['pitcher_id'] + ".html"

        with open(filename, 'wb') as f:
            f.write(response.body)
```

#### Compression ####

Once the file system was converted to pitcher-based structure, another script was created to generate individual csv files for each pitcher. e.g.

Below is the core code snippet used. The actual python script used to re-structure the files can be found [here](https://github.com/azampagl/ai-ml-pitch/blob/master/wrangle/compress.py).

```python
def compress():
    """Main execution."""

    # Determine command line arguments.
    try:
        rawopts, _ = getopt.getopt(sys.argv[1:], 'i:o:')
    except getopt.GetoptError:
        usage()
        sys.exit(2)

    opts = {}

    # Process each command line argument.
    for o, a in rawopts:
        opts[o[1]] = a

    # The following arguments are required in all cases.
    for opt in ['i', 'o']:
        if not opt in opts:
            usage()
            sys.exit(2)

    # Make sure the output directory exists.
    if not os.path.exists(opts['o']):
        os.makedirs(opts['o'])

    # Traverse the root folder that contains sub folders
    #  that represent each pitcher.
    for root, dirs, _ in os.walk(opts['i']):
        # Traverse each folder in the root.
        for pid in dirs:
            outfile = os.path.join(opts['o'], pid + ".csv")

            # Check if this pitcher was already processed.
            if os.path.isfile(outfile):
                continue

            for proot, _, files in os.walk(os.path.join(root, pid)):
                try:
                    # Read in the first game for this pitcher.
                    with open(os.path.join(proot, files[0]), 'r') as f:
                        df = pd.read_html(f.read(), header=0)[0]
                    # Read in the subsequent games and append to the
                    #  running DataFrame.
                    for file in files[1:]:
                        with open(os.path.join(proot, file), 'r') as f:
                            df = df.append(pd.read_html(f.read(), header=0)[0])
                    # Save to disk as a csv file.
                    df.to_csv(outfile)
                except ValueError:
                    print("Error processing " + pid)
                    continue
```

### Pitching Strategy During a Game ### 

Analyzing a pitcher's strategy over the course of a game could provide insight, effecting an opposing coach decisions. 

We approached this question by looking at the occurrence of different pitch types for a pitcher over the course of a game averaged over his career. It is important to understand that we define "time" by the number of pitches during a game. We use the number of pitches as a measurement because there could be a wide range of pitches in a cetrain inning or certain block of time (e.g. 40 pitches in inning 3 and 9 pitches in inning 6). 

Below is the source code used to analyze the column entitled "mlbam_pitch_name", which contained the pitch type (e.g. Fastball, Curve Ball, Sinker, etc.). We generated a bar graph for each pitch type per game over the course of a pitcher's career. The Y axis was the number of occurrences of that particular pitch type and the X axis was the "window" of pitches (e.g. pitches 1-10, 10-20, 20-30).

```python
def analyze_column_over_time(filename, column_name, pitches_per_window, out_dir):
    """
    Loads in the pitcher data and visualizes the desired column over time.
    """

    # Read in the dataframe.
    df = pd.read_csv(filename)

    # Determine all the possible distinct types in the column.
    column_types = df[column_name] \
                    .value_counts().reset_index()['index'].tolist()

    # Initialize the dictionary that keeps track of the average
    #  occurrences for a particular column value per sliding window.
    type_avgs = dict.fromkeys(column_types, None)
    for k, v in type_avgs.items():
        type_avgs[k] = []

    # Calculate the total occurrences of a type
    # per sliding window per game for a pitchers entire career.
    type_by_range = {}
    for gid, gid_group in df.groupby('gid'):
        # Find the total number of pitches for this game.
        num_pitches = gid_group.shape[0]

        # Evaluate the pitches in a sliding window (pitches_per_group).
        i = pitches_per_window
        while (i < num_pitches):
            # Gather the pitches in this sliding range.
            pitches = gid_group.iloc[i - pitches_per_window:i]
            # Determine the counts per pitch type.
            type_counts = Counter(
                dict(pitches[column_name].value_counts())
            )

            # Add the counts to a dictionary for the current range (window).
            if (not i in type_by_range):
                type_by_range[i] = [type_counts]
            else:
                type_by_range[i].append(type_counts)

            # Increment the sliding range.
            i += pitches_per_window

    # For each sliding window, determine the average per type.
    for pitch_window, type_counts in iter(sorted(type_by_range.iteritems())):
        type_counts_total = Counter()
        # Add up the pitch counts for the current window.
        for type_count in type_counts:
            type_counts_total += type_count

        # Convert back into a dictionary to calculate averages.
        for k, v in dict(type_counts_total).iteritems():
            type_avgs[k].append(float(v) / len(type_counts))

    # Find the pitch that occurred in the most amount of sliding windows
    #  per game and extend all the other pitches so they are the same size
    #  This is done so the visualization generates properly.
    max_windows = 0
    for k, v in type_avgs.iteritems():
        max_windows = max(max_windows, len(v))
    for k, v in type_avgs.items():
        v.extend([0.0] * (max_windows - len(v)))

    ind = np.arange(max_windows)
    width = 0.15

    # Visualize each pitch type individually so we can see how
    #  the pitch type occurrences change over time (number of pitches)
    i = 0
    for column_type, type_counts in type_avgs.iteritems():
        fig = plt.figure(figsize=(20, 10))
        ax = fig.add_subplot(111)
        ax.bar(ind, type_counts)
        ax.set_xticks(ind)
        ax.set_xticklabels(
            [str(i * pitches_per_window) + "-" + str((i + 1) * pitches_per_window) \
            for i in xrange(max_windows)]
        )
        ax.set_title(
            "\"" + str(column_type) + "\"" + \
            " Over Time for Column " + column_name)
        ax.set_xlabel("Pitch Range")
        ax.set_ylabel("Number of Occurrences")

        fig.savefig(
            os.path.join(
                out_dir,
                str(int(df['pitcher_id'][0])) + "-" + column_name + "-" + \
                str(column_type) + "-n" + str(pitches_per_window) + ".png"
            )
        )
        i += 1
```

The following function will be used to generate summary statistics for a particular matchup versus Felix Hernandez

```python
# A function used to compute several summary statistics detailing the percentage of pitches Hernandez has thrown against the 
# player with corresponding batter_id.
# output: boolean param for printed output
# Returns: a list containing '% Changeups','% Curveballs','% 4-Seam Fastballs','% Cutter','% Sinker',
#   '% Slider','Avg Speed of pitch','Avg Pitches per AB', '% of pitches that are fastballs','% of first pitch fastballs' for
#   the given batter

def matchup_analysis(batter_id, all_samples=False,output=False):
    #look at the subset of pitch Hernandez vs Kinsler
    
    matchup=hern_edited[hern_edited['batter_id']==batter_id]

    #sanity check
    if(output): print len(matchup)


    by_pitch_type=matchup.groupby('mlbam_pitch_name')
    if(output): print("Breakdown of all pitches by percentage: ")
    #store how many pitches this pitcher has, to be used later
    diff_pitches=0
    pitch_dict={}   #going to use this dictionary for visuals
    for pitch_type,subset in by_pitch_type:
        if(output): print(str(pitch_type)+" : "+str(len(subset)*1.0/len(matchup)*100.00))
        diff_pitches+=1
        pitch_dict[str(pitch_type)]=[]     #just store a container

    if(output): print
    avg_speed=matchup['start_speed'].mean()
    if(output): print("Average speed:" +str(avg_speed))
    avg_pitch_per_ab=matchup[matchup['ab_count']==1]['ab_total'].mean()
    if(output): print ("Avg pitches per at-bat: "+str(avg_pitch_per_ab))
    if(output):  print
    
    fb=sum(matchup['fastball_binary'])*1.0/len(matchup)
    if(output): print("Percentage of pitches that are fastballs:"+str(fb*100))
    avg_speed_fb=matchup[matchup['fastball_binary']==1]['start_speed'].mean()
    if(output): print("Avg speed of fastballs: "+str(avg_speed_fb))
    if(output): print
    if(output): print("Percentage of pitches that aren't fastballs:"+str((1-fb)*100))
    avg_speed_nfb=matchup[matchup['fastball_binary']!=1]['start_speed'].mean()
    if(output): print("Avg speed of non-fastballs: "+str(avg_speed_nfb))
    if(output): print

    #store the subset of first pitches in the at-bat
    first_pitch=matchup[matchup['ab_count']==1]
    first_pitch_fb=sum(first_pitch['fastball_binary'])*1.0/len(first_pitch)
    if(output): print("Percentage of first pitch fastballs:"+ str(first_pitch_fb*100.0))
    
    group_year=matchup.groupby('year')
    
    fb_year=[]
    nfb_year=[]
    years=[]
    for year,group in group_year:
        years.append(year)
        #want to generate year by year statistics to see if any change over time
        fb_frac=sum(group['fastball_binary'])*1.0/len(group)
        fb_year.append(fb_frac)
        nfb_year.append(1-fb_frac)
    
    
    summary_vec=[]
    ##build a list of values to be returned. Eventually can be used to populate a table to compare batters
    tot=len(matchup)
    # % changeups
    summary_vec.append(len(matchup[matchup['mlbam_pitch_name']=='CH'])*100.0/tot)
    # % curveballs
    summary_vec.append(len(matchup[matchup['mlbam_pitch_name']=='CU'])*100.0/tot)
    # % four-seam fastballs, FA and FF
    temp=(matchup['mlbam_pitch_name']=='FA')|(matchup['mlbam_pitch_name']=='FF')
    summary_vec.append(len(matchup[temp])*100.0/tot)
    # % cutter
    summary_vec.append(len(matchup[matchup['mlbam_pitch_name']=='FC'])*100.0/tot)
    # % sinker, SI and FS
    temp= (matchup['mlbam_pitch_name']=='SI')|( matchup['mlbam_pitch_name']=='FS')
    summary_vec.append(len(matchup[temp])*100.0/tot)
    # % slider
    summary_vec.append(len(matchup[matchup['mlbam_pitch_name']=='SL'])*100.0/tot)
    #avg speed of pitch
    summary_vec.append(avg_speed)
    # avg pitch per at-bat
    summary_vec.append(avg_pitch_per_ab)
    # % of pitches that are fastballs
    summary_vec.append(fb*100)
    # % first pitch fastballs
    summary_vec.append(first_pitch_fb*100)
    
    return summary_vec
   


```

#### Comment on function below: ####

This function returns a dataframe whose rows are the returned values of matchup_analysis() for all batters who have faced more than 70 pitches thrown by Felix Hernandez. We are only considering batters with more than 70 pitches faced against Hernandez so to not distort percentages. We could have weighted percentages by the number of pitches faced, but there were enough batters with # pitches > 70 that it wasn't necessary.

```python
#function to return a dataframe composed of the summary statistics from matchup_analysis for all batters face by Hernandez
# Returns: a dataframe where each row contains the matchup statistics returned from matchup_analysis()
def general_summary():
   
    batter_grouping=hern_edited.groupby('batter_id')
    batter_dict = dict(iter(batter_grouping))

    # only considering batters with more than 70 pitches faced against Hernandez so to not distort percentages. Could have
    # weighted percentages by the number of pitches faced, but there were enough batters with # pitches > 70 that it wasn't
    # necessary
    batters=[x for x in batter_dict.keys() if len(batter_dict[x])>70 ]

    summary_stats=[]
    for batter in batters:
        summary_stats.append(matchup_analysis(batter))
        
    cols=['% Changeups','% Curveballs','% 4-Seam Fastballs','% Cutter','% Sinker','% Slider','Avg Speed of pitch','Avg Pitches per AB',
                               '% of pitches that are fastballs','% of first pitch fastballs' ]
    batter_df=pd.DataFrame(summary_stats, columns=cols)
    
    return batter_df
```

### Visual Analysis Via Radar Plot: ###

In order to have another view of the deviation in tendencies agains the six batters selected, a radar plot was cosntructed. The percentages for each type of pitch and other stats was scaled by the average so that the radar plot for the average tendency is uniforrm in all directions. The average tendency is marked in black and the tendencies against each of the players is marked in different colors as seen in the legend. It can be seen for example, that batter Kendrick (in grey) received a higher percentage of cutter and slider pitches than the rest of the group.  


```python
#define radar plot

def radar_factory(num_vars, frame='circle'):
    # calculate evenly-spaced axis angles
    theta = 2*np.pi * np.linspace(0, 1-1./num_vars, num_vars)
    # rotate theta such that the first axis is at the top
    theta += np.pi/2

    def draw_poly_patch(self):
        verts = unit_poly_verts(theta)
        return plt.Polygon(verts, closed=True, edgecolor='k')

    def draw_circle_patch(self):

        return plt.Circle((0.5, 0.5), 0.5)

    patch_dict = {'polygon': draw_poly_patch, 'circle': draw_circle_patch}
    if frame not in patch_dict:
        raise ValueError('unknown value for `frame`: %s' % frame)

    class RadarAxes(PolarAxes):

        name = 'radar'

        RESOLUTION = 1

        draw_patch = patch_dict[frame]

        def fill(self, *args, **kwargs):
            """Override fill so that line is closed by default"""
            closed = kwargs.pop('closed', True)
            return super(RadarAxes, self).fill(closed=closed, *args, **kwargs)

        def plot(self, *args, **kwargs):
            """Override plot so that line is closed by default"""
            lines = super(RadarAxes, self).plot(*args, **kwargs)
            for line in lines:
                self._close_line(line)

        def _close_line(self, line):
            x, y = line.get_data()

            if x[0] != x[-1]:
                x = np.concatenate((x, [x[0]]))
                y = np.concatenate((y, [y[0]]))
                line.set_data(x, y)

        def set_varlabels(self, labels):
            self.set_thetagrids(theta * 180/np.pi, labels)

        def _gen_axes_patch(self):
            return self.draw_patch()

        def _gen_axes_spines(self):
            if frame == 'circle':
                return PolarAxes._gen_axes_spines(self)

            # spine_type must be 'left', 'right', 'top', 'bottom', or `circle`.
            spine_type = 'circle'
            verts = unit_poly_verts(theta)
            # close off polygon by repeating first vertex
            verts.append(verts[0])
            path = Path(verts)

            spine = Spine(self, spine_type, path)
            spine.set_transform(self.transAxes)
            return {'polar': spine}

    register_projection(RadarAxes)
    return theta


def unit_poly_verts(theta):
    x0, y0, r = [0.5] * 3
    verts = [(r*np.cos(t) + x0, r*np.sin(t) + y0) for t in theta]
    return verts
```

###Feature Importance###

Next, we use a dimension reduction method based on random forests to find "important" dimensions of the dataset to focus on when predicting the type of pitch. Random forests allow us to compute a heuristic for determining how "important" a feature is in predicting a target. This heuristic measures the change in prediction accuracy if we take a given feature and permute (scramble) it across the datapoints in the training set. The more the accuracy drops when the feature is permuted, the more "important" we can conclude the feature is. Importance can be a useful way to select a small number of features for visualization.
This feature importance analysis allows us to understand which factors have bearing on the pitching category.

The function below will be used to generate and plot features by importance for a particular pitch category using a random forest classifier 

```python
def plotFeatureImportance(X,Y,X2,title):
    # Build a forest and compute the feature importances
    forest = RandomForestClassifier(n_estimators=15, random_state=0)

    forest.fit(X, Y)
    importances = forest.feature_importances_
    std = np.std([tree.feature_importances_ for tree in forest.estimators_],axis=0)
    indices = np.argsort(importances)[::-1]

    # Print the feature ranking
    print("Feature importance ranking:")

    for f in range(len(importances)):
        #print("%d. feature %d --(%f)" % (f + 1, indices[f], importances[indices[f]]))
        i = indices[f]
        print f+1,'. feature', indices[f], '(', X2[[i]].columns[0],') -', round(importances[indices[f]],2)

    # Plot the feature importances of the forest
    plt.figure(figsize=(15,5))
    plt.title("Feature importance for: "+title)
    plt.bar(range(len(importances)), importances[indices],color="r", yerr=std[indices], align="center")
    plt.xticks(range(len(importances)-1), X2[indices].columns)
    plt.xlim([-1, 10])
    plt.show()
```

The first feature importance analysis we are interested in is for Fastballs. We define the target column and dataset to be used in the classifier below

### Establishing Priors ###

Following the ideas of Guttag in his [paper](http://www.sloansportsconference.com/wp-content/uploads/2012/02/98-Predicting-the-Next-Pitch_updated.pdf) on pitch prediction, we established a 'prior' on the probability Hernandez will throw a fastball against a given batter_id. To build our prior, we looked at the percentage of pitches each batter_id faced against Hernandez in 2011 and 2012 that were classified as a type of fastball. The logic behind this is that we can use the past percentage of pitch types against a particular batter to inform us about how a pitcher may pitch against that batter in the future. This 'prior' may be distorted for batters who have not faced Hernandez many times. Guttag's paper suggests a way to generate a shrunken prior towards Hernandez's career tendencies for batters with low support. This was not something we had time to investigate, but it's a logical next step.

### 'New' functions ###

We recycled the matchup_analysis and gen_summary functions from earlier to be used on a particular subset of my data. The goal here is to generate batter-pitcher priors for the percentage of pitches that are fastballs based on data from 2011 and 2012. These priors will be used as features in our predictive model while using 2013, 2014 data as our test set.

```python


# Similar to matchup_analysis() except to be used on particular years. To be used to build pitcher-batter priors for our predictive model
def matchup_plus(ds,batter_id, all_samples=False,output=False,num_samp=False):
    #look at the subset of pitch Hernandez vs Kinsler
    
    matchup=ds[ds['batter_id']==batter_id]

    #sanity check
    if(output): print len(matchup)


    by_pitch_type=matchup.groupby('mlbam_pitch_name')
    if(output): print("Breakdown of all pitches by percentage: ")
    #store how many pitches this pitcher has, to be used later
    diff_pitches=0
    pitch_dict={}   #going to use this dictionary for visuals
    for pitch_type,subset in by_pitch_type:
        if(output): print(str(pitch_type)+" : "+str(len(subset)*1.0/len(matchup)*100.00))
        diff_pitches+=1
        pitch_dict[str(pitch_type)]=[]     #just store a container

    if(output): print
    avg_speed=matchup['start_speed'].mean()
    if(output): print("Average speed:" +str(avg_speed))
    avg_pitch_per_ab=matchup[matchup['ab_count']==1]['ab_total'].mean()
    if(output): print ("Avg pitches per at-bat: "+str(avg_pitch_per_ab))
    if(output):  print
    
    fb=sum(matchup['fastball_binary'])*1.0/len(matchup)
    if(output): print("Percentage of pitches that are fastballs:"+str(fb*100))
    avg_speed_fb=matchup[matchup['fastball_binary']==1]['start_speed'].mean()
    if(output): print("Avg speed of fastballs: "+str(avg_speed_fb))
    if(output): print
    if(output): print("Percentage of pitches that aren't fastballs:"+str((1-fb)*100))
    avg_speed_nfb=matchup[matchup['fastball_binary']!=1]['start_speed'].mean()
    if(output): print("Avg speed of non-fastballs: "+str(avg_speed_nfb))
    if(output): print

    #store the subset of first pitches in the at-bat
    first_pitch=matchup[matchup['ab_count']==1]
    first_pitch_fb=sum(first_pitch['fastball_binary'])*1.0/len(first_pitch)
    if(output): print("Percentage of first pitch fastballs:"+ str(first_pitch_fb*100.0))
    
    group_year=matchup.groupby('year')
    
    fb_year=[]
    nfb_year=[]
    years=[]
    for year,group in group_year:
        years.append(year)
        #want to generate year by year statistics to see if any change over time
        fb_frac=sum(group['fastball_binary'])*1.0/len(group)
        fb_year.append(fb_frac)
        nfb_year.append(1-fb_frac)
    
    
    summary_vec=[]
    ##build a list of values to be returned. Eventually can be used to populate a table to compare batters
    tot=len(matchup)
    # % changeups
    summary_vec.append(len(matchup[matchup['mlbam_pitch_name']=='CH'])*100.0/tot)
    # % curveballs
    summary_vec.append(len(matchup[matchup['mlbam_pitch_name']=='CU'])*100.0/tot)
    # % four-seam fastballs, FA and FF
    temp=(matchup['mlbam_pitch_name']=='FA')|(matchup['mlbam_pitch_name']=='FF')
    summary_vec.append(len(matchup[temp])*100.0/tot)
    # % cutter
    summary_vec.append(len(matchup[matchup['mlbam_pitch_name']=='FC'])*100.0/tot)
    # % sinker, SI and FS
    temp= (matchup['mlbam_pitch_name']=='SI')|( matchup['mlbam_pitch_name']=='FS')
    summary_vec.append(len(matchup[temp])*100.0/tot)
    # % slider
    summary_vec.append(len(matchup[matchup['mlbam_pitch_name']=='SL'])*100.0/tot)
    #avg speed of pitch
    summary_vec.append(avg_speed)
    # avg pitch per at-bat
    summary_vec.append(avg_pitch_per_ab)
    # % of pitches that are fastballs
    summary_vec.append(fb*100)
    # % first pitch fastballs
    summary_vec.append(first_pitch_fb*100)
    
    if(num_samp==True):
        summary_vec.append(len(matchup))
        summary_vec.append(batter_id)
        
    return summary_vec
   
```

```python
#Exactly like general summary except not excluding batters with <70 pitches seen. Using this to create pitcher-batter priors
#for prediction. Interested only in fastball/not-fastball. 
# Input: A subset of the hernandez dataframe that contains a 'batter_id' column
# Output: a dataframe with 3 columns,'% of pitches that are fastballs','num_pitches','batter_id' generated from year_ds

def general_summary_2(year_ds):
   
    batter_grouping=year_ds.groupby('batter_id')
    batter_dict = dict(iter(batter_grouping))

    # only considering batters with more than 70 pitches faced against Hernandez so to not distort percentages. Could have
    # weighted percentages by the number of pitches faced, but there were enough batters with # pitches > 70 that it wasn't
    # necessary

    summary_stats=[]
    for batter in batter_dict:
        summary_stats.append(matchup_plus(year_ds,batter,num_samp=True))
        
    cols=['% Changeups','% Curveballs','% 4-Seam Fastballs','% Cutter','% Sinker','% Slider','Avg Speed of pitch','Avg Pitches per AB',
                               '% of pitches that are fastballs','% of first pitch fastballs','num_pitches','batter_id' ]
    batter_df=pd.DataFrame(summary_stats, columns=cols)
    
    batter_df=batter_df[['% of pitches that are fastballs','num_pitches','batter_id']]
    
    return batter_df

fastball_summary_by_batter=general_summary_2(hern_2011_12)
        
```


The function below looks up the priors for a column of batter_id's, batter_coumn, from another argument 'summary' which is the output of general_summary_2() function above


```python

#Inputs: 
#  batter_column: a pandas Series that contains batter_id values
#  summary: a pandas dataframe that contains 3 columns; % of pitches that are fastballs, number of pitches seen, batter_id 
# output:
#  A series where each row of the output corresponds to the % of pitches that are fastballs from summary for the corresponding batter_id.
#  If the batter_id is not found in summary, the mean value of % of pitches that are fastballs in summary is returned
# Example: 
#   batter_column=[1, 2, 5, 1, 3]; summary=[ [.65, 40, 1] , [.75, 200, 2] , [.60, 130, 5] ]
#   output= [.65, .75, .60, .65, .65]
def generate_priors(batter_column, summary):
    batter_list=batter_column.astype(list)
    sum_list=summary['% of pitches that are fastballs'].values
    
    #store the mean fastball % to be used for batter_id not found in summary
    mean_val= summary['% of pitches that are fastballs'].mean()
    
    output=[]
    
    for batter in batter_list:
        #assign mean value to those batters who haven't faced hernandez yet
        if(len(summary[summary['batter_id']==batter])==0):
            output.append(mean_val)
        else:
            ind=summary[summary['batter_id']==batter].index
            output.append(sum_list[ind][0]/100.0)
            
    return(output)
            
```
